```python
# ==========================================
# Maincrafts Task 3
# Titanic Mini Exploratory Data Analysis
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Improve Plot Style
sns.set_style("whitegrid")

# ===============================
# Load Dataset
# ===============================

# Use the URL for the correct Titanic dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

print("="*60)
print("First 5 Rows")
print(df.head())

print("\nDataset Shape:", df.shape)

print("\nColumns")
print(df.columns)

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

print("\nMissing Values")
print(df.isnull().sum())

# ===============================
# Data Cleaning
# ===============================

# Fill missing Age values with Mean
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Fill missing Embarked values with Mode
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin Column
if "Cabin" in df.columns:
    df.drop(columns=["Cabin"], inplace=True)

print("\nMissing Values After Cleaning")
print(df.isnull().sum())

# ===============================
# Feature Engineering
# ===============================

# Family Size
df["FamilySize"] = df["SibSp"] + df["Parch"]

# Age Groups
bins = [0,12,18,35,60,100]
labels = ["Child","Teen","Adult","Middle Age","Senior"]

df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=bins,
    labels=labels
)

# ===============================
# Analysis 1
# Survival by Age Group
# ===============================

print("\n==============================")
print("Survival Rate by Age Group")
print("==============================")

age_group = df.groupby("AgeGroup")["Survived"].mean()

print(age_group)

# ===============================
# Analysis 2
# Survival by Embarkation Port
# ===============================

print("\n==============================")
print("Survival Rate by Embarkation Port")
print("==============================")

embarked = df.groupby("Embarked")["Survived"].mean()

print(embarked)

# ===============================
# Analysis 3
# Survival by Family Size
# ===============================

print("\n==============================")
print("Survival Rate by Family Size")
print("==============================")

family = df.groupby("FamilySize")["Survived"].mean()

print(family)

# ===============================
# Visualization 1
# Age Distribution Histogram
# ===============================

plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Age",
    bins=30,
    kde=True,
    color="royalblue"
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()

# ===============================
# Visualization 2
# Correlation Heatmap
# ===============================

plt.figure(figsize=(10,8))

numeric = df.select_dtypes(include=np.number)

sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()

# ===============================
# Visualization 3
# Survival by Family Size
# ===============================

plt.figure(figsize=(10,5))

sns.barplot(
    x=family.index,
    y=family.values,
    palette="viridis"
)

plt.title("Survival Rate by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")

plt.show()

# ===============================
# Visualization 4
# Survival by Age Group
# ===============================

plt.figure(figsize=(8,5))

sns.barplot(
    x=age_group.index,
    y=age_group.values,
    palette="Set2"
)

plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")

plt.show()

# ===============================
# Visualization 5
# Survival by Embarkation Port
# ===============================

plt.figure(figsize=(6,5))

sns.barplot(
    x=embarked.index,
    y=embarked.values,
    palette="pastel"
)

plt.title("Survival Rate by Embarkation Port")
plt.xlabel("Embarkation Port")
plt.ylabel("Survival Rate")

plt.show()

# ===============================
# Visualization 6
# Survival Count
# ===============================

plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="Survived",
    palette="Set1"
)

plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Passengers")

plt.show()

# ===============================
# Key Insights
# ===============================

print("\n" + "="*60)
print("KEY INSIGHTS")
print("="*60)

print("1. Missing Age values were replaced with Mean.")
print("2. Missing Embarked values were replaced with Mode.")
print("3. Cabin column was removed because it contained many missing values.")
print("4. FamilySize feature was created using SibSp + Parch.")
print("5. Adults generally showed a better survival rate.")
print("6. Survival rate differed across Embarkation Ports.")
print("7. Family size influenced passenger survival.")
print("8. Correlation heatmap shows relationships among numerical features.")

print("\nTask 3 Completed Successfully.")
```
